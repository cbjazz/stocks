# Histogram & Bestfit

```python
from scipy.stats import norm

fig, ax = plt.subplots()

(mu, sigma) = norm.fit(df_samsung_elec['s_close'])
n, bins, patches = ax.hist(df_samsung_elec['s_close'], 60, normed=1, alpha=0.75)

y = mlab.normpdf(bins, mu, sigma)
ax.plot(bins, y, 'r--', linewidth=2)

plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
fig.tight_layout()
plt.show()
```
